```python
import random
import pandas as pd
```

```python
quiz_df = pd.read_json("questions.json")
quiz_df.head()
```

```python
def run_quiz(questions_df, num_questions, pass_percentage):
    selected_questions = questions_df.sample(n=num_questions, replace=False)
    score = 0

    print("Welcome to the Quiz!")

    for i, row in selected_questions.iterrows():
        question_text = row['question']

        num_correct_answers = len(row['correct_answers'])
        if num_correct_answers > 1:
            question_text += f" (Select {num_correct_answers} answers)"

        print(f"\nQuestion {i + 1}: {question_text}")

        for j, option in enumerate(row['answer_options'], start=1):
            print(f"{j}. {option}")

        user_choices = input(
            "Enter your answers (comma-separated): ").split(',')
        user_choices = [choice.strip() for choice in user_choices]

        if all(choice in row['correct_answers'] for choice in user_choices) and len(user_choices) == num_correct_answers:
            score += 1

    print("\nQuiz completed!")
    print(f"Your score: {score}/{num_questions}")

    if (score / num_questions) >= pass_percentage:
        print("Congratulations! You passed.")
    else:
        print("Sorry, you did not pass.")


run_quiz(quiz_df, num_questions=20, pass_percentage=0.8)
```
